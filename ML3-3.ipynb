```python
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1})
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assume the data is stored in a variable called 'data'
# Example: data = pd.read_csv('breast-cancer.csv')  # Uncomment to load the dataset

# X: All columns except the second one (features)
X = data.iloc[:, [i for i in range(data.shape[1]) if i != 1]]

# y: The second column is the target variable (label)
y = data.iloc[:, 1]

# Combine features (X) and target (y) into a single DataFrame for correlation analysis
data_with_target = pd.concat([X, y], axis=1)

# Calculate the correlation matrix to understand relationships between variables
correlation_matrix = data_with_target.corr()

# Plot the correlation matrix as a heatmap for visual interpretation
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Extract correlations of each feature with the target variable (last column in the combined DataFrame)
target_correlation = correlation_matrix.iloc[:-1, -1]  # Last column is the target; exclude it from features

# Print the features sorted by their correlation with the target, from highest to lowest
print("Feature correlations with the target:")
print(target_correlation.sort_values(ascending=False))
```

```python
from sklearn.model_selection import train_test_split

# Select only the top 10 features based on correlation
selected_features = [
    'concave points_worst',
    'perimeter_worst',
    'concave points_mean',
    'radius_worst',
    'perimeter_mean',
    'area_worst',
    'radius_mean',
    'area_mean',
    'concavity_mean',
    'concavity_worst'
]

```

```python
# Separate features (X) and target (y)
X = data[selected_features]  # Only use selected features
y = data['diagnosis']        # Target column


from sklearn.model_selection import train_test_split

# X: features, y: labels (target)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,          # 20% for testing
    random_state=42,        # for reproducibility
    shuffle=True,           # always shuffle unless data is time-series
                 # preserves class distribution in train/test sets
)
```

```python
# Optional: Print shapes to verify
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
for col in selected_features:
    Q1 = eda_data[col].quantile(0.25)
    Q3 = eda_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # delete outliers
    eda_data = eda_data[(eda_data[col] >= lower_bound) & (eda_data[col] <= upper_bound)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=eda_data[selected_features])
plt.xticks(rotation=45)
plt.title('Boxplots After Removing Outliers')
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import roc_curve, auc
```

```python
kernels = [ 'rbf', 'linear', 'poly']
results = {}
```

```python
for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)

    results[kernel] = {
        'accuracy': accuracy,
        'conf_matrix': conf_matrix,
        'class_report': class_report
    }

    print(f"SVM kernel = {kernel}")
    print(f"(Accuracy): {accuracy:.4f}")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)
```

```python
best_kernel = max(results, key=lambda k: results[k]['accuracy'])
print("best kernel")
print(f"Kernel: {best_kernel} | accuracy: {results[best_kernel]['accuracy']:.4f}")
```

```python
svm_model = SVC(kernel='linear', probability=True)
svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_test)

print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm))
```

```python
cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues')
plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob_svm = svm_model.predict_proba(X_test)[:, 1]
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)
auc_svm = auc(fpr_svm, tpr_svm)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
nn_model = Sequential([
    Dense(16, input_shape=(X_train.shape[1],), activation='relu'),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = nn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=0)
```

```python
loss, accuracy = nn_model.evaluate(X_test, y_test, verbose=0)
print(f"\nNeural Network Accuracy: {accuracy}")

y_pred_nn = (nn_model.predict(X_test) > 0.5).astype("int32")
print("Neural Network Classification Report:")
print(classification_report(y_test, y_pred_nn))
```

```python
cm_nn = confusion_matrix(y_test, y_pred_nn)
sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Greens')
plt.title("Neural Network Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob_nn = nn_model.predict(X_test).ravel()
fpr_nn, tpr_nn, _ = roc_curve(y_test, y_prob_nn)
auc_nn = auc(fpr_nn, tpr_nn)
```

```python
plt.plot(fpr_svm, tpr_svm, label=f"SVM (AUC = {auc_svm:.2f})")
plt.plot(fpr_nn, tpr_nn, label=f"Neural Network (AUC = {auc_nn:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(y_true, y_pred, model_name):
    print(f"--- {model_name} ---")
    print(f"Accuracy : {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred, average='weighted'):.4f}")
    print(f"Recall   : {recall_score(y_true, y_pred, average='weighted'):.4f}")
    print(f"F1 Score : {f1_score(y_true, y_pred, average='weighted'):.4f}\n")
# Use y_pred_svm and y_pred_nn, which were assigned previously
evaluate_model(y_test, y_pred_svm, "SVM")
evaluate_model(y_test, y_pred_nn, "Neural Network")
```
